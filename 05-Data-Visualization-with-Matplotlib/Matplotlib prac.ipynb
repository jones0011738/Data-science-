```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
# Functional
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# object oriented
fig= plt.figure()
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig= plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.3,0.3])
axes1.plot(x,y,'r')
axes2.plot(y,x,'b')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
# axes.plot(x,y)

# for current_ax in axes:
#     plt.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('First')
axes[1].plot(x,y)
axes[1].set_title('Second')
```

```python
# fig=plt.figure(figsize=(3,2))
# ax=fig.add_axes([0,0,1,1])
# ax.plot(x,y)
# or 
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(5,5))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X_SQUARED')
ax.plot(x,x**3,label='X_CUBED')
ax.legend()
```

```python
fig= plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=2,
#         (or lw),
        alpha=0.5,linestyle='--',
#         (or ls)
        marker='o',markersize=20,
        markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
